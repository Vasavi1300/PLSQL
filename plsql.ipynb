declare
    cursor c_emp is
        select employee_id, last_name from employees;
 
    TYPE c_emp_rec IS RECORD (
        employee_id employees.employee_id%TYPE
        ,last_name employees.last_name%TYPE
    );
    TYPE c_emp_typ is table of c_emp_rec INDEX BY BINARY_INTEGER;
     c_emp_tbl c_emp_typ;
begin
    open c_emp;
    loop
        fetch c_emp bulk collect into c_emp_tbl Limit 50;
        exit when c_emp%notfound;
 
        for i in 1..c_emp_tbl.count
        LOOP
            update employees
            set email = first_name || c_emp_tbl(i).last_name || '@deloitte.com'
            where employee_id = c_emp_tbl(i).employee_id;
        end loop;
    end loop;
    close c_emp;
end;

CREATE OR REPLACE PACKAGE BODY employee_user_name_pkg AS
    PROCEDURE create_user_name_prc (
        p_last_name IN employees.last_name%TYPE,
        l_user_name OUT employees.user_name%TYPE
    ) IS
        cnt NUMBER;
    BEGIN
        SELECT
            reverse(substr(reverse(p_last_name),
                           1,
                           3))
        INTO l_user_name
        FROM
            dual;

        SELECT
            COUNT(last_name)
        INTO cnt
        FROM
            employees
        WHERE
            user_name = l_user_name;

        IF cnt > 1 THEN
            l_user_name := l_user_name || '01';
        END IF;
    END create_user_name_prc;

    PROCEDURE update_user_name_prc IS
        CURSOR emp_cur IS
        SELECT
            employee_id,
            last_name
        FROM
            employees;

        l_user_name employees.last_name%TYPE;
    BEGIN
        FOR emp_cur_rec IN emp_cur LOOP
            --create_user_name_prc(emp_cur_rec.last_name, l_user_name);
            UPDATE employees
            SET
                user_name = l_user_name
            WHERE
                employee_id = emp_cur_rec.employee_id;

            COMMIT;
        END LOOP;
    END update_user_name_prc;
END employee_user_name_pkg;

create or replace package body employees_pkg
as
	procedure create_user_name(p_last_name employees.last_name%type
								,l_user_name out employees.user_name%type)
	IS
		V_COUNT NUMBER;
		v_temp employees.user_name%type;
    begin
		select reverse(substr(reverse(p_last_name),1,4)) into l_user_name from dual;
		v_temp := l_user_name;
		LOOP
			select count(*) into v_count from employees where reverse(substr(reverse(p_last_name),1,4))=v_temp;
			if v_count > 0
			then
				v_temp:=l_user_name ||str(v_count);
			else
				exit;
			end if;
		END LOOP;
	end;

    procedure update_username as
        l_user_name employees.user_name%type;
        cursor c1 is 
            select last_name,employee_id from employees;
    begin
        for c1_rec in c1
        loop
			create_user_name(c1_rec.last_name,l_user_name);
			update employees set user_name=l_user_name where employee_id=c1_rec.employee_id;
			commit;
        end loop;
    end update_username;
end employees_pkg;
